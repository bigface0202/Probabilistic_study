```python
# ベクトルxを [] の区間で作成
n = np.linspace(-100.0, 100.0, 100)

# 平均loc, 標準偏差scaleの正規分布における、xの確率を求める
for pp in range(3):
    mu = pm.Uniform("percep_avg", 0,50)
    tau = pm.Uniform("percep_std", 0,10)
    p = []
    for i in range(len(n)):
        p.append(st.norm.pdf(x=n[i], loc=mu.value, scale=tau.value))
    plt.plot(n, p)
# 乱数－確率 の特性を散布図で表し、標準正規分布のグラフを作成
```

```python
#5N与えられた時の誤差を正規分布と仮定してそのμとσをmcmcによって推定
mu1 = pm.Uniform("mu1",-10,10)
sig1 = pm.Uniform("sig1",0,5)
obs1 = pm.Normal("obs1", mu1, 1/sig1**2, value = sabun['5'], observed = True) #pm.Normalはシグマ（分散）ではなく，タウ（確度）なので気をつける
mcmc = pm.MCMC([mu1, sig1])
mcmc.sample(25000,5000)
```

```python
help(pm.Normal)
```

```python

mu1_sample = mcmc.trace("mu1")[:,None]
sig1_sample = mcmc.trace("sig1")[:,None]

figsize(12.5,10)

plt.hist(mu1_sample, bins = 30, alpha = 0.85, color="red", normed = True)
plt.hist(sig1_sample, bins = 30, alpha = 0.85, color="green", normed = True)
plt.xlim(-2,5)
```

```python
def norm_out(x,mu, sigma):
    return 1/(np.sqrt(2*np.pi*sigma))*np.exp(-(x-mu)**2/(2*sigma))
```

```python
mu1_sample
```

```python
x = np.linspace(-5,5,50)[:,None]
p = norm_out(x.T, mu1_sample, sig1_sample)

mean_prob = p.mean(axis=0)

plt.plot(x,mean_prob)
```

```python
len(mean_prob)
```

```python
from scipy.stats.mstats import mquantiles

qs = mquantiles(p, [0.025, 0.975], axis = 0)
plt.fill_between(x[:,0], *qs, alpha = 0.7, color = "blue")

plt.plot(x[:,0], qs[0], label="95% CI", color = "#7A68A6", alpha = 0.7)
plt.plot(x,mean_prob, ls="--", color="k")
```
