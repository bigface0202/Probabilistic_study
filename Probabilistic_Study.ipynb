```python
#5N与えられた時の誤差を正規分布と仮定してそのμとσをmcmcによって推定
mu1 = pm.Uniform("mu1",-20,10)
sig1 = pm.Uniform("sig1",0,5)
obs1 = pm.Normal("obs1", mu1, 1/sig1**2, value = sabun['40'], observed = True) #pm.Normalはシグマ（分散）ではなく，タウ（確度）なので気をつける
model = pm.Model([obs1,mu1,sig1])
mcmc = pm.MCMC(model)
mcmc.sample(25000,5000)
```

```python
help(pm.Normal)
```

```python

mu1_sample = mcmc.trace("mu1")[:,None]
sig1_sample = mcmc.trace("sig1")[:,None]

figsize(12.5,10)

plt.hist(mu1_sample, bins = 30, alpha = 0.85, color="red", normed = True)
plt.hist(sig1_sample, bins = 30, alpha = 0.85, color="green", normed = True)
plt.xlim(-20,5)
```

```python
def norm_out(x,mu, sigma):
    return 1/(np.sqrt(2*np.pi*sigma))*np.exp(-(x-mu)**2/(2*sigma))
```

```python
mu1_sample
```

```python
x = np.linspace(-20,5,50)[:,None]
p = norm_out(x.T, mu1_sample, sig1_sample)

mean_prob = p.mean(axis=0)

plt.plot(x,mean_prob)
```

```python
#muは筋活動度によって決まる
def _mu(alpha):
    if alpha < 0.03:
        return pm.Uniform("mu", -1,1).value
    else:
        return -389 * alpha + 12
```

```python
#各筋活動度α毎に誤差を算出
def error_distribution(activation):
    sig_list = []
    mu_list = []
    for act in activation['act']:
        mu1 = _mu(act)
        sig1 = pm.Uniform("sig1",0,5)
        obs1 = pm.Normal("obs1", mu1, 1/sig1**2, value = sabun2[str(act)], observed = True) #pm.Normalはシグマ（分散）ではなく，タウ（確度）なので気をつける
        model = pm.Model([obs1,mu1,sig1])
        mcmc = pm.MCMC(model)
        mcmc.sample(25000,5000)
        #mu1_sample = mcmc.trace("mu1")[:,None]
        sig1_sample = mcmc.trace("sig1")[:,None]
        x = np.linspace(-20,5,50)[:,None]
        p = norm_out(x.T, mu1, sig1_sample)
        sig_list.append(sig1_sample.mean())
        mu_list.append(mu1)
        mean_prob = p.mean(axis=0)
        plt.plot(x,mean_prob)
        plt.figure()
    return sig_list, mu_list
```

```python
sig_tes, mu_tes = error_distribution(act_std_df)
```

```python
for i, act in enumerate(act_std_df['act']):
    mu = _mu(act)
    sig = sig_tes[i]
    error = pm.rnormal(mu,sig,100)
    x = np.zeros(len(error))
    x[:] = act
    if(i < 1):
        x_ = x
        error_ = error
    else:
        x_ = np.r_[x_,x]
        error_ = np.r_[error_,error]
```

```python
def Perception_Force(Fa, mu, sig):
    for n in range(len(mu)):
        if(n < 13):
            degree = 10 * n
        elif(n == 13):
            degree = 120
        else:
            degree = 120 - 10 * (n-13)
        error = pm.rnormal(mu[n],sig[n],100)
        Fa_ = np.zeros(len(error))
        Fa_[:] = Fa[n]
        degree_ = np.zeros(len(error))
        degree_[:] = degree
        if(n < 1):
            _Fa_ = Fa_
            error_ = error
            _degree_ = degree_
        else:
            _Fa_ = np.r_[_Fa_,Fa_]
            error_ = np.r_[error_,error]
            _degree_ = np.r_[_degree_,degree_]
    Fp = _Fa_ + error_
    return pd.DataFrame({'Fa':_Fa_,
                        'Fp':Fp,
                       'Degree':_degree_})
```
